#### Ejercicio 16. Calcula la media y la varianza de muestra para las siguientes variables: precio, habitaciones, superficie y baños (debes usar el subconjunto obtenido del Ejercicio 14) (★★★)

Imprime ambos valores por cada variable.

```python
# TODO
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
ds_cinturon_sur = ds[ds["level5"].isin(cinturon_sur)]
variables = ["price", "rooms", "bathrooms", "surface"]

results = pd.DataFrame(columns=["Variable", "Media", "Varianza"])

temp_list = []

for var in variables:
    media = ds_cinturon_sur[var].mean()
    varianza = ds_cinturon_sur[var].var()
    temp_list.append({"Variable": var, "Media": media, "Varianza": varianza})

results = pd.concat([results, pd.DataFrame(temp_list)], ignore_index=True)

print(results)
```

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
ds_cinturon_sur = ds[ds["level5"].isin(cinturon_sur)]

def normalize_min_max(column):
    return (column - column.min()) / (column.max() - column.min())
ds_cinturon_sur["price_normalized"] = ds_cinturon_sur.groupby("level5")["price"].transform(normalize_min_max)

colors = ["red", "blue", "yellow", "orange"]

for i, population in enumerate (cinturon_sur):
    data_normalized = ds_cinturon_sur.loc[ds_cinturon_sur["level5"] == population]["price_normalized"]
    plt.hist(data_normalized, bins = 10, alpha = 0.5, label = population, color = colors[i])

plt.title ("Histograma normalizado")
plt.legend()
plt.xlabel("Frecuencia")
plt.ylabel("Precio normalizado")
plt.show()
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# TODO
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
ds_cinturon_sur = ds[ds["level5"].isin(cinturon_sur)].copy()
ds_cinturon_sur["pps"] = ds_cinturon_sur["price"] / ds_cinturon_sur["surface"]
ds_getafe_alcorcon = ds_cinturon_sur[ds_cinturon_sur["level5"].isin(["Getafe", "Alcorcón"])]
pps_getafe = ds_getafe_alcorcon[ds_getafe_alcorcon["level5"] == "Getafe"]["pps"].mean()
pps_alcorcon = ds_getafe_alcorcon[ds_getafe_alcorcon["level5"] == "Alcorcón"]["pps"].mean()

print(f"Precio por metro cuadrado promedio en Getafe: {pps_getafe:.2f} eur/m2")
print(f"Precio por metro cuadrado promedio en Alcorcón: {pps_alcorcon:.2f} eur/m2")
print("Podemos observar que en Alcorcón el precio por metro cuadrado es más alto que en Getafe.")
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
ds_cinturon_sur = ds[ds["level5"].isin(cinturon_sur)].copy()
ds_cinturon_sur["pps"] = ds_cinturon_sur["price"] / ds_cinturon_sur["surface"]

fig, axs = plt.subplots (2, 2, figsize = (10, 12))
for i, poblacion in enumerate (cinturon_sur):
    data = ds_cinturon_sur[ds_cinturon_sur["level5"] == poblacion]
    ax = axs[i//2, i % 2]
    ax.scatter(data["surface"], data["pps"], color = "red", alpha = 0.7)
    ax.set_title(poblacion)
    ax.set_xlabel("Surface")
    ax.set_ylabel("Price per square")
    ax.set_xlim(0, 300)
    ax.set_ylim(1000, 3000)

plt.tight_layout()
plt.show()
```
